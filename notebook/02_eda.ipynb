```python
# Target variable distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='study_status', data=df, color="#2E86C1")
plt.title('Study Status Distribution') #RECRUITING: 1, NOT_YET_RECRUITING: 0

df['study_status'].value_counts(normalize=True)
```

### Univariate Analysis

```python
#Numerical Features
# Study Duration
sns.histplot(df['study_duration_days'], bins=30, kde=True, color='cornflowerblue')
plt.title('Distribution of Study Duration (Days)')
plt.xlabel('Study Duration (days)')
plt.show()
```

```python
# Enrollment
sns.histplot(df['enrollment'], bins=30, kde=True, color='darkorange')
plt.title('Distribution of Enrollment Numbers')
plt.xlabel('Enrollment')
plt.show()
```

```python
#Categorical Features
# Study Type
sns.countplot(x='study_type', data=df, hue='study_type', palette='Set1')
plt.title('Study Type Distribution')
plt.show()

# Top 10 Study Designs
top_designs = df['study_design'].value_counts().head(10)
top_designs.plot(kind='bar', color='skyblue')
plt.title('Top 10 Study Designs')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
#Conditions
top_conditions = df['conditions'].value_counts().head(10)
top_conditions.plot(kind='bar', color='teal')
plt.title('Top 10 Medical Conditions in Trials')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
bool_cols = [
    'sex_all', 'sex_female', 'sex_male',
    'has_child', 'has_adult', 'has_older_adult',
    'phase1', 'phase2', 'phase3'
]

df[bool_cols].sum().sort_values().plot(kind='bar', color='mediumseagreen')
plt.title('Boolean Features Distribution')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
funder_cols = [col for col in df.columns if col.startswith('funder_')]

df[funder_cols].sum().sort_values().plot(kind='bar', color='salmon')
plt.title('Funder Type Distribution')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

### Bivariate Analysis

```python
# Enrollment vs Study Status
sns.boxplot(x='study_status', y='enrollment',hue='study_status', data=df, palette='Set3')
plt.title('Enrollment by Study Status')
plt.xlabel('Study Status')
plt.ylabel('Enrollment')
plt.show()
```

```python
# Study Duration vs Study Status
sns.boxplot(x='study_status', y='study_duration_days', hue='study_status', data=df, palette='Set3')
plt.title('Study Duration by Study Status')
plt.xlabel('Study Status')
plt.ylabel('Study Duration (days)')
plt.show()
```

```python
# Study Type vs Study Status
sns.countplot(x='study_type', hue='study_status', data=df, palette='coolwarm')
plt.title('Study Type vs Study Status')
plt.xticks(rotation=0)
plt.show()
```

```python
# Phase vs Study Status
phase_cols = ['phase1', 'phase2', 'phase3']
for col in phase_cols:
    sns.barplot(x=col, y='study_status', hue=col, data=df, palette='viridis')
    plt.title(f'{col.upper()} vs Study Status')
    plt.ylabel('Proportion of Recruiting')
    plt.show()
```

```python
# Correlation heatmap for numerical features
# Compute correlation
corr = df.select_dtypes(include=['int64', 'float64', 'bool']).corr()

# Mask for upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Create heatmap
sns.heatmap(
    corr,
    mask=mask,
    cmap='coolwarm',
    annot=True,
    fmt=".2f",        
    linewidths=.5,
    cbar_kws={"shrink": .7},
    annot_kws={"size": 9}
)

plt.title("Improved Correlation Heatmap", fontsize=16, pad=20)
plt.xticks(rotation=90, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Select numeric/boolean columns only
num_df = df.select_dtypes(include=['int64', 'float64', 'bool'])

# Calculate correlations with target
corr_with_target = num_df.corr()['study_status'].sort_values(ascending=False).drop('study_status')

# Plot
plt.figure(figsize=(8, 10))
sns.heatmap(corr_with_target.to_frame(), annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title("Correlation of Features with Study Status", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
sample_df = df.sample(500, random_state=42)
sns.pairplot(sample_df[['study_status', 'enrollment', 'study_duration_days']], hue='study_status')
plt.suptitle("Pairplot of Key Features vs Study Status", y=1.02)
plt.show()
```

```python
sample_df = df.sample(500, random_state=1)

# Pairplot with all features except the target 
features = sample_df.select_dtypes(include=['number', 'category']).columns.tolist()

if 'study_status' not in features:
    features.append('study_status')

# Create pairplot
pair_plot = sns.pairplot(
    sample_df[features],
    hue='study_status',                 
    diag_kind='kde',                    
    markers='o',                      
    plot_kws={'alpha': 0.7, 's': 40},  
    corner=False,                      
    palette='Set2',                    
    height=2.5                        
)

plt.suptitle("Pairplot of All Features vs Study Status", y=1.02, fontsize=16)
plt.show()
```
